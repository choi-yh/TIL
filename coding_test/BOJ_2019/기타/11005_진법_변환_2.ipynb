```
"""
B진수 만드는 규칙대로 만들고 10이 넘어갈 경우 chr로 변환해주면 된다.
"""

import sys
n, b = map(int, input().split())

result = ''
while n // b > 0:
    residue = n % b
    n //= b
    if residue >= 10:
        residue = chr(residue + 55)
    result = str(residue) + result

if n >= 10:
    n = chr(n + 55)
result = str(n) + result

print(result)
```
